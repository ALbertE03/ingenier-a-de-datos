```python
import pandas as pd
import os
```

```python
INITIAL_DATA_DIR = os.path.join(os.getcwd(), "initial_data")
```

```python
def path(*kwargs):
    return os.path.join(*kwargs)
```

```python
list_files = os.listdir(INITIAL_DATA_DIR)
l =[]
for file in list_files:
    if not l:
        l.append(pd.read_csv(path(INITIAL_DATA_DIR,file)))
    else:
        csv = pd.read_csv(path(INITIAL_DATA_DIR,file))
        df = pd.concat([l[-1],csv],axis=0)
        l.append(df)
        
        
df = l[-1]
```

```python
df.sort_values(by="Report.Date",ascending=True,inplace=True)
```
